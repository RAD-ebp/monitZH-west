# Read ZST Location Data

Anpassen der ZST Koordinaten auf WGS84

```python
import pandas as pd
from pyproj import Proj, transform  # Import Proj and transform from pyproj
```

```python
# Umrechnung von LV95 in WGS84 
def lv95_to_latlon(x_lv95, y_lv95):
    lon, lat = transform(lv95, wgs84, x_lv95, y_lv95)
    return lat, lon
```

```python
# read marker data
marker_data = pd.read_excel(r'C:\Users\RAD\EBP\CH_P_223609 - Monitoring ZH West - Monitoring Zürich West\40_BEARBEITUNG\03_Zählstellen\Zählstellen Attributtabelle.xlsx')

# Festlegen der Koordinatensysteme
lv95 = Proj(init='epsg:2056')
wgs84 = Proj(init='epsg:4326')  # EPSG code for WGS84

# nur wenn, Werte vorhanden anwenden.
marker_data[['WGS84_lat', 'WGS84_lon']] = marker_data.apply(lambda row: pd.Series(lv95_to_latlon(row['Koordinaten E (LV95)'], row['Koordinaten N (LV95)'])), axis=1)

#save df
marker_data.to_excel(r'C:\Users\RAD\EBP\CH_P_223609 - Monitoring ZH West - Monitoring Zürich West\40_BEARBEITUNG\03_Zählstellen\Zählstellen Attributtabelle.xlsx', index=False)

# display
marker_data
```
